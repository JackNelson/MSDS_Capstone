```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats

class NumSummaryStats:
    
    def __init__(self, filepath, var):
        self.filepath = filepath
        self.var = var
        
        df = pd.read_csv(filepath,lineterminator='\n')
        dup = df.shape[0] - df.drop_duplicates().shape[0]
        print ("Duplicates: "+str(dup))
        #print("kstest for Laplace: "+str((stats.kstest(np.log(df[var][df[var] > 0]), 'laplace'))))
        print ("%s: Zero = %4d Nonzero = %7d" % (var, df[var][df[var] == 0].count(), df[var][df[var] > 0].count()))
        sns.distplot(np.log(df[var][df[var] > 0]), fit = stats.laplace) 
```
