# CoinAPI Tweet Pull

### Instructions:
- import the CoinAPI.py program
- initialize the CoinAPI object class with given API Key
- use the loopHistTweets() function to sample a full day's worth of information (100 tweets/hour every hour)
- export out file to a .txt (formatted as a .csv) using the saveTweetsText() function
- every 24 hours 4 days can be sampled using a key from the free version of coinAPI.io (100 requests/day / 24 requests/loopHistTweets() function
- files are saved under data/coin_tweets
- to check how many requests remaining in a given 24 hour period call api.request_limits 
- <u>NOTE</u>: request limit attributes are only accurately updated after a request is made to coinAPI.io, when object is initialized it is defaulted to 100 requests remaining assuming the end user is in a fresh 24 hour period

```python
# run CoinAPI.py
%run -i 'CoinAPI.py'

# intialize CoinAPI
api = CoinAPI('XXXXXXXX-XXXX-XXXX-XXXX-XXXXXXXXXXXX')

# gather tweets
list_of_tweets = api.loopHistTweets(time_start = '2017-11-01T00:00:00')

# save file in coin_tweets folder
outfile_name = '20171101.txt'
api.saveTweetsText(list_of_tweets, outfile_name)
```
